## If-Statements with User-Defined Functions

```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

### Busiest day of the week

```python
# Finding  two busiest days of the week 
busiestdays = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0 or value == 1 :
   busiestdays.append("Busiest days")
  elif value == 3 or value == 4:
   busiestdays.append("Least busy")
  else:
   busiestdays.append("Regularly busy")
```

```python
df_orders_products_merged['busiest_days'] = busiestdays
```

```python
df_orders_products_merged['busiest_days'].value_counts(dropna = False)
```

### Busiest hours of the day

```python
# Finding busiest hour of the day
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
orders = []


for value in df_orders_products_merged["order_hour_of_day"]:
  if value >=9 and value <=17 :
   orders.append("Most orders")
  elif value >=2 and value<=5:
   orders.append("Fewest orders")
  else:
   orders.append("Average orders")
```

```python
df_orders_products_merged['busiest_period_of_day'] = orders
```

```python
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna=False)
```
