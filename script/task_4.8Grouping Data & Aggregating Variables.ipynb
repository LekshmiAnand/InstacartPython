### Finding Loyal Customers

```python
#creating a loyalty flag for existing customers using the transform() and loc() functions
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

#### Loyalty Flag 

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','max','min']})
```

### Finding types of Spenders 

```python
#Create a spending flag for each user based on the average price across all their orders 
ords_prods_merge['avg_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge[['user_id', 'avg_price', 'prices']].head(60)
```

#### Spender Flag

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10 , 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

### Frequent versus non-frequent customers

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'order_frequency']].head(60)


```

#### Frequency flag 

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] >= 20 , 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] <= 20) & (ords_prods_merge['order_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10 , 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
